```python
def med(a,b):
 m,n=len(a),len(b)
 dp=[[i+j if i*j==0 else 0 for j in range(n+1)] for i in range(m+1)]
 for i in range(1,m+1):
  for j in range(1,n+1):
   dp[i][j]=dp[i-1][j-1] if a[i-1]==b[j-1] else 1+min(dp[i-1][j],dp[i][j-1],dp[i-1][j-1])
 return dp[m][n]
tests=[("kitten","sitting"),("flaw","lawn"),("intention","execution"),("apple","aple"),("recieve","receive")]
for x,y in tests:
 print(f"MED({x},{y}) =",med(x,y))
```

```python
class TD:
 def __init__(s,x): s.x,s.i=x,0
 def m(s,c):
  if s.i<len(s.x) and s.x[s.i]==c: s.i+=1; return True
  return False
 def parse(s):
  return s.m('a') and (s.m('b') or True) and s.m('c') and s.i==len(s.x)

print("Top Down:", TD("abc").parse())


class BU:
 def __init__(s,x): s.x=list(x); s.st=[]
 def parse(s):
  while s.x or len(s.st)>1:
   if s.x: s.st.append(s.x.pop(0))
   if ''.join(s.st[-2:])=='bc': s.st[-2:]=['A']
   elif ''.join(s.st[-2:])=='aA': s.st[-2:]=['S']
  return s.st==['S']

print("Bottom Up:", BU("abc").parse())
```

```python
data=[("fun couple love love","Comedy"),("fast furious shoot","Action"),
      ("couple fly fast fun fun","Comedy"),("furious shoot shoot fun","Action"),
      ("fly fast shoot love","Action")]

wc={"Comedy":{}, "Action":{}}
for t,c in data:
 for w in t.split(): wc[c][w]=wc[c].get(w,0)+1

def predict(t):
 return max(wc, key=lambda c: sum(wc[c].get(w,0) for w in t.split()))

t="fast couple shoot fly"
print("Test Doc:",t)
print("Predicted Class:",predict(t))
```

```python
import nltk
from nltk.corpus import brown, inaugural, reuters, udhr
from nltk import FreqDist, ConditionalFreqDist
from nltk.tag import pos_tag

nltk.download('brown'); nltk.download('inaugural'); nltk.download('reuters'); nltk.download('udhr')
nltk.download('punkt'); nltk.download('averaged_perceptron_tagger')

print("Brown:", brown.words()[:10])
print("Inaugural:", inaugural.words()[:10])
print("Reuters:", reuters.words()[:10])
print("UDHR:", udhr.words('English-Latin1')[:10])

fd = FreqDist(brown.words())
print("Freq:", fd.most_common(5))

cfd = ConditionalFreqDist((genre, word) for genre in brown.categories() for word in brown.words(categories=genre))
print("CFD sample:", list(cfd.conditions())[:3])

tags = pos_tag(brown.words()[:20])
print("Tagged:", tags)

nouns = [w for w,t in tags if t.startswith('NN')]
print("Nouns:", nouns)
```

```python

import nltk
from nltk.corpus import brown
from nltk import FreqDist
from nltk.tag import pos_tag

nltk.download('brown')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

# sample words
words = brown.words()[:20]
print("Words:", words)

# frequency
print("Freq:", FreqDist(words).most_common(3))

# POS tagging
tags = pos_tag(words)
print("Tags:", tags)

# nouns
print("Nouns:", [w for w,t in tags if t.startswith('NN')])
```

```python
#Lemmatization
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

l=WordNetLemmatizer()
words=["running","cars","better","playing"]

for w in words:
 print(w,"->",l.lemmatize(w))
```

```python
# Sentiment Analysis
from textblob import TextBlob

text="I love this movie. It is amazing"

result=TextBlob(text)

print("Polarity:",result.sentiment.polarity)
print("Subjectivity:",result.sentiment.subjectivity)
```
