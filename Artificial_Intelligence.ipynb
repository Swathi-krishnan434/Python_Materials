```python
# AI Anomaly Detection
from sklearn.ensemble import IsolationForest

data=[[10],[11],[12],[10],[11],[100]]

model=IsolationForest(random_state=42)
result=model.fit_predict(data)

for x,r in zip(data,result):
 if r==-1:
  print(x,"Anomaly")
 else:
  print(x,"Normal")
```

```python
# AI Spam Detection
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

msg=["Win money now",
     "Meeting at 10",
     "Claim free prize",
     "Complete your assignment"]

label=["Spam","Not Spam","Spam","Not Spam"]

v=CountVectorizer()
x=v.fit_transform(msg)

model=MultinomialNB()
model.fit(x,label)

text=input("Enter Message: ")

print("Prediction:",
model.predict(v.transform([text]))[0])
```

```python
# AI Fake News Detection
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

text=["Government announces new policy",
      "Aliens landed on earth yesterday",
      "Scientists discover new medicine",
      "Moon made of cheese"]

label=["Real","Fake","Real","Fake"]

v=CountVectorizer()
x=v.fit_transform(text)

model=MultinomialNB()
model.fit(x,label)

news=input("Enter News: ")

print("Prediction:",
model.predict(v.transform([news]))[0])
```
